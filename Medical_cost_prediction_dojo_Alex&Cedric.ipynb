```python
#Define target and features
X = df_ml[['age', 'sex', 'bmi', 'children', 'smoker', 'region_northeast', 'region_northwest', 'region_southeast', 'region_southwest']]
y = df_ml['charges']
```

```python
#Scale
scaler = StandardScaler()#MinMaxScaler()
X_scaled = scaler.fit_transform(X)
pd.DataFrame(X_scaled).describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state = 49)

model = DecisionTreeRegressor(max_depth=5, min_samples_leaf=11)
model.fit(X_train,y_train)

#model.predict(X_test)

print("\nScore for the Train dataset :", model.score(X_train, y_train))
print("Score for the Test dataset :", model.score(X_test, y_test))
print("mean square error",mean_squared_error(y_test,model.predict(X_test)))
```

```python
cross_val_score(DecisionTreeRegressor(max_depth=5, min_samples_leaf=11),X_train,y_train,cv=6)
```

```python
parameters = {'max_depth':range(1,10), 'min_samples_leaf':range(1,16), 'min_samples_split':range(1,15)}
dt = DecisionTreeRegressor()
gs = GridSearchCV(dt, parameters)
gs.fit(X_train, y_train)
# type(gs.best_estimator_)
dt = gs.best_estimator_
dt.fit(X_train, y_train)
result = {}
result['score'] = dt.score(X_test, y_test)
result['mean_squared_error'] = mean_squared_error(y_test,dt.predict(X_test))
print("The best score is:",gs.best_score_)
print("The best parameters are:",gs.best_params_)
result
```

```python
regression_list = [{'model': LinearRegression(), 'parameters': {'n_jobs': range(2,21)}, 'results': {}},
                       
                       {'model': KNeighborsRegressor(), 'parameters': {'n_neighbors' :range(2,20),'weights':['uniform', 'distance'],'leaf_size':range(1,20)}, 'results': {}},
                       {'model': MLPRegressor(), 'parameters': {'activation':['identity', 'logistic', 'tanh', 'relu'],'solver':['lbfgs', 'sgd', 'adam']}, 'results': {}},
                   {'model': DecisionTreeRegressor(), 'parameters': {'max_depth':range(1,10), 'min_samples_leaf':range(1,16), 'min_samples_split':range(1,15)}, 'results': {}}]
#{'model': SVR(), 'parameters': {'kernel':['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']}, 'results': {}},
#{'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'] ,  'max_depth': range(1,20)}
```

```python
def run_the_model(regression, X_train, X_test, y_train, y_test):
    #print('new variable')
    parameters = regression['parameters']
    model = regression['model']
    #print(model)
    gs = GridSearchCV(model, parameters)
    gs.fit(X_train, y_train)
    # type(gs.best_estimator_)
    model = gs.best_estimator_
    #print(model)
    model.fit(X_train, y_train)
    regression['model'] = model
    result = {}
    result['score'] = model.score(X_test, y_test)
    result['mean_squared_error'] = mean_squared_error(y_test,model.predict(X_test))
    #result['precision'] = precision_score(y_test, model.predict(X_test))
    #result['recall'] = recall_score(y_test, model.predict(X_test))
    regression['results'] = result
    return regression
```

```python
result_regression = []
for regressor in regression_list:
    result_regression += [run_the_model(regressor, X_train, X_test, y_train, y_test)]
```
